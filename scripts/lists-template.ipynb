### Populate and print a list

```python
beatles = ["John", "Paul", "George", "Ringo", "Stuart"]
# index; John is at index 0, so on & so forth


#print the whole list
print(beatles)
```

### Calculate and print the length of a list

```python
# calculate number of beatles
number_of_beatles = len(beatles)

# print number of beatles
print(f"There are {number_of_beatles} members of the Beatles.")
```

### Print an item in a list 
   
   _see PowerPoint slides_

```python
#remind oursleves of the list
print(beatles)

# print the first beatle (John)
print(beatles[0])

# many ways to print the last Beatle (Stuart)
print(beatles[-1]) #Stuart
print(beatles[4]) #Stuart
print(beatles[len(beatles)-1]) #Stuart

#many ways to print George
print(beatles[2]) #george index from start
print(beatles[-3]) #george index from end
```

### Remove an item from a list

```python
# Remove the last item in the list (bye, Stuart)
stu = beatles.pop()

# What happened to the Beatles?
print(stu)
print(beatles)
```

```python

# extend() modifies a list in place, it doesn't return a value (i.e., a new list)
jackson5.extend(destiny)

#here's what it's doing
# jackson5 = jackson5 + destiny

print(f"Here's the extended Jackson5: {jackson5}")
```

### Growing a list with `append()`

### Stings as lists

- strings are just a list of characters

- this means you can index into a string, the same way you would a list

```python
name = 'beyonce'
print(name[0], name[-1])
```

```python
# here's a gene with structure EXONintronEXON
gene = 'ATGCCAGTttccggaaGTTCTAA'

# extract the first exon (i.e., the first string of capital letters: 'ATGCCAGT')
print(f'the first exon: {gene[0:8]}')

# extract the start codon (the first three characters of the gene)
print(gene[0:3])

# extract the stop codon (the last three characters)
print(gene[-3:])

# extract the intron
print(gene[8:16])


```

```python
# split our gene into a list
gene_as_list = list(gene)
print(gene_as_list)
```
